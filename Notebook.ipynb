initialisation and loading flights data

```python
import gzip
import pandas as pd
from datetime import datetime

print("extracting csv.gz ...", end="")
filename = "flightlist_20200401_20200430.csv.gz"
f = gzip.open(filename, mode='rt')
print("done")

start_time = datetime.now()
print("Start opening csv using pandas (",start_time,")")
df = pd.read_csv(f)
end_time = datetime.now()
print("Complete opening csv using pandas (",end_time,")")
elapsed_time = (end_time - start_time)
print("Elapsed time: ", str(elapsed_time.seconds), "second(s)")
print()
df.head()
```

load airports data

```python
df_airports = pd.read_csv("airports.csv")
df_airports.head()
```

removing all rows with null value for origin

```python
print("Rows with null value for origin - before removal: ", df.origin.isna().sum())
df.dropna(subset=['origin'], inplace=True)
print("Rows with null value for origin - after removal: ", df.origin.isna().sum())
df.head()
```

removing all rows with null value for destination

```python
print("Rows with null value for destination - before removal: ", df.destination.isna().sum())
df.dropna(subset=['destination'], inplace=True)
print("Rows with null value for destination - after removal: ", df.destination.isna().sum())
df.head()
```

drop all unnecessary columns

```python
columns_to_keep = ['callsign', 'origin', 'destination', 'day']
df = df[columns_to_keep]
df.head()
```

merge the flights and airports dataframe to link gps_code and origin

```python
columns_to_keep += ['iso_region']
df_flightlist_airports = pd.merge(df, df_airports, left_on='origin', right_on='gps_code', how='left')[columns_to_keep]
df_flightlist_airports.rename(columns={'iso_region':'orig_region'}, inplace=True)
df_flightlist_airports.head()
```

merge the flights and airports dataframe to link gps_code and destination. Keep the `dest_region` and `type` columns (`type` column to be renamed as `dest_type`)

```python
columns_to_keep = list(df_flightlist_airports.columns)
columns_to_keep += ['iso_region','type']
df_flightlist_airports = pd.merge(df_flightlist_airports, df_airports, left_on='destination', right_on='gps_code', how='left')[columns_to_keep]
df_flightlist_airports.rename(columns={'iso_region':'dest_region', 'type':'dest_type'}, inplace=True)
df_flightlist_airports.head()
```

Since the earlier merging is done using left join, it is important to make sure that all records containing null value are removed

```python
print("Number of records - before null removal: ", str(len(df_flightlist_airports)))
df_flightlist_airports.dropna(subset=['orig_region','dest_region'], inplace=True)
print("Number of records - after null removal: ", str(len(df_flightlist_airports)))
```

Only keep the rows with `dest_region` started with "US-"

```python
df_flightlist_airports = df_flightlist_airports[df_flightlist_airports['dest_region'].str.startswith("US-")]
df_flightlist_airports.head()
```

classify each row as either international or domestic flight based on the `orig_region` value. This category is put into a new column, namely `flight_type`

```python
df_flightlist_airports['flight_type'] = df_flightlist_airports['orig_region'].apply(lambda x: "Domestic" if str(x).startswith("US-") else "International")
df_flightlist_airports.head()
```

remove the time from the `day` column, resulting the `day` column contains only the date

```python
df_flightlist_airports['day'] = df_flightlist_airports['day'].apply(lambda x: str(x).split()[0])
df_flightlist_airports.sample(5)
```

exporting the final result into a .csv.gz file

```python
df_flightlist_airports.to_csv("flights_airports.csv.gz", index=False, sep="|", compression="gzip")
```
